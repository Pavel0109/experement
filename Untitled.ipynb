```python
import pandas as pd
import numpy as np
import math
import plotly.graph_objects as go
```

```python
kof = 1.5
h = 100
v_rabbit = 10
v_fox = v_rabbit * kof

delta_t = 0.1

rabbit_trajectory = [[]]
fox_trajectory = [[]]

rabbit_trajectory=[[0],[0],[h]]
fox_trajectory=[[0],[0],[0]]
ras_mass = []
```

```python
i = 0
t = 0
alpha = math.pi/2
ras = h
ras_mass.append(ras)
while ras > 1:
    
    ras = math.sqrt(abs((rabbit_trajectory[1][i])-fox_trajectory[1][i])**2+
                abs((rabbit_trajectory[2][i]-fox_trajectory[2][i]))**2)
    t = t + delta_t
    i = i + 1
    
    rabbit_trajectory[0].append(t)
    rabbit_trajectory[1].append(t*v_rabbit)
    rabbit_trajectory[2].append(h)
    
    fox_trajectory[0].append(t)
    fox_trajectory[1].append(fox_trajectory[1][i-1] + v_fox * (-fox_trajectory[1][i-1]+rabbit_trajectory[1][i-1])/ras*delta_t)
    fox_trajectory[2].append(fox_trajectory[2][i-1] + v_fox * (-fox_trajectory[2][i-1]+rabbit_trajectory[2][i-1])/ras*delta_t)
    ras_mass.append(ras)
    print(i,'fox =',fox_trajectory[1][i],fox_trajectory[2][i])
    print(i,'rabbit = ',rabbit_trajectory[1][i],rabbit_trajectory[2][i])
    print(i,'ras=',ras)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = rabbit_trajectory[1],y=rabbit_trajectory[2],
                    mode='markers',
                    name='rabbit'))
fig.add_trace(go.Scatter(x = fox_trajectory[1],y=fox_trajectory[2],
                    mode='markers',
                    name='fox'))
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = rabbit_trajectory[0],y=ras_mass,
                    mode='markers',
                    name='rabbit'))
```
