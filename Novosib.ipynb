$f = CS\rho V^2/2$

```python
import numpy as np
import math
import plotly.graph_objects as go
```

```python
C = 0.4
S = 0.3
ro = 1.26
v0 = 10
alpha = math.pi/3
g = 9.8
h0 = 0
x = [0]
y = [h0]
t = [0]
dt = 0.01
m = 1
v=v0
v_y = [v0*math.sin(alpha)]
v_x = [v0*math.cos(alpha)]
def f (C,S,ro,v):
    return C*S*ro*(v0**2)/2
#v_x,v_y,f(C,S,ro,v)
v_x[0]
```

```python
i = 0
while y[i] >= 0:
    a_x = -f(C,S,ro,v_x[i])
    a_y = -g-f(C,S,ro,v_y[i])
    
    x.append(x[i]+dt*v_x[i])
    y.append(y[i]+dt*v_y[i])
    t.append(t[i]+dt)
    
    v_y.append(v_y[i]+a_y*dt)
    v_x.append(v_x[i]+a_x*dt)
    print(x[i],y[i])
    i=i+1
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = x,y=y,
                    mode='markers'))

fig.show()
```

```python
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x = t, y=v_x, mode='markers', name = 'v_x'))
fig2.add_trace(go.Scatter(x = t, y=v_y, mode='markers', name = 'v_y'))

fig2.show()
```
